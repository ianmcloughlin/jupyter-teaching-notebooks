### Square root of 2

***

When $ x^2 = 2 $, what is $x$?

$ x^2 = 2 \Rightarrow x^2 - 2 = 0$

$ f(x) = x^2 - 2 $

When $f(x) = 0$, what is $x$?

<br>

```python
# The function above.
def f(x):
    return x**2 - 2.0
```

```python
# Create a plot.
fig, ax = plt.subplots()

# An array of values for plotting f(x).
xvals = np.linspace(-3.0, 3.0, 1000)

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='lightgrey', linestyle='--')

# Plot the x's and y's.
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Legend.
ax.legend();
```

### Derivative

***

$ f(x) = x^2 - 2 = x^2 - 2 x^0$

$ f^\prime(x) = 2x^1 - 0(2 x^{-1}) = 2x $

$ y = mx +c = 2x + 0$

<br>

```python
# The derivative of f(x).
def df(x):
    return 2.0 * x
```

```python
# Create a plot.
fig, ax = plt.subplots()

# An array of values for plotting f(x).
xvals = np.linspace(-3.0, 3.0, 1000)

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='lightgrey', linestyle='--')

# Plot f(x).
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Plot f'(x.)
ax.plot(xvals, df(xvals), color='C0', label='Derivative')

# Legend.
ax.legend();
```

```python
# The slope of f(x) when x=3.0.
df(3.0)
```

```python
# The value of f(x) when x=3.0.
f(3.0)
```

### Tangents

***

Equation of a line from a point and the slope:

$y - y_1 = m(x - x_1)$

$y = m(x - x_1) + y_1$

<br>

```python
# An array of values for the plot.
xvals = np.linspace(0.0, 6.0, 1000)
```

```python
# The tangent of f(x) at f(3.0).
tan = (df(3.0) * (xvals - 3.0)) + f(3.0)
```

```python
# Create a plot.
fig, ax = plt.subplots()

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='lightgrey', linestyle='--')

# Plot f(x).
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Plot tan.
ax.plot(xvals, tan, color='C3', label='Tangent')

# Legend.
ax.legend();
```

```python
# Generic tangent calculation.
def tan_at_x1(x, x1):
    return (df(x1) * (x - x1)) + f(x1)
```

```python
# Create a plot.
fig, ax = plt.subplots()

# An array of values for the plot.
xvals = np.linspace(-3.0, 3.0, 1000)

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='k', alpha=0.2, linestyle='--')

# Plot f(x).
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Plot f'(x).
ax.plot(xvals, df(xvals), color='C0', label='Derivative')

# Plot tan at x=1.0.
ax.plot(xvals, tan_at_x1(xvals, 1.0), color='C3', label='Tangent')

# Legend.
ax.legend();
```

### Integral

***

$ f(x) = x^2 - 2 = x^2 - 2 x^0$

$ F(x) = \int\,f\,dx = \frac{x^3}{3} - 2 \frac{x^1}{1} = \frac{1}{3} x^3 - 2 x$

Technically we are ignoring the constant ($+c$).

<br>

```python
# Integral in Python.
def F(x):
    return ((x**3) / 3.0) - (2.0 * x)
```

```python
# Create a plot.
fig, ax = plt.subplots()

# An array of values for plotting f(x).
xvals = np.linspace(-3.0, 3.0, 1000)

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='k', alpha=0.2, linestyle='--')

# Plot f(x).
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Plot f'(x).
ax.plot(xvals, df(xvals), color='C0', label='Derivative')

# Plot F(x).
ax.plot(xvals, F(xvals), color='C2', label='Integral')

# Legend.
ax.legend();
```

### Gradient Descent

***

```python
# Create a plot.
fig, ax = plt.subplots()

# An array of values for plotting f(x).
xvals = np.linspace(-3.0, 3.0, 1000)

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='k', alpha=0.2, linestyle='--')

# Plot f(x).
ax.plot(xvals, f(xvals), color='C1', label='Original')

# Plot F(x).
ax.plot(xvals, F(xvals), color='C2', label='Integral')

# Legend.
ax.legend();
```

```python
# An initial guess at the square root.
guess = 3.0
```

```python
# Value of f at current guess.
f(guess)
```

```python
# The value of F at current guess.
F(guess)
```

```python
# The maximum number of iterations.
max_iter = 200

# Some measure of the required accuracy.
accuracy = 0.05

# The little bit to move x by.
learning_rate = 0.01

# An initial guess at the square root.
guess = 3.0

# Set iterations count to 0.
i = 0

# Slowly decrease x, and watch the value of F.
while abs(f(guess)) > accuracy:
    # Print the guess and it's F value.
    if i % 10 == 0:
        print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")
    # Adjust the current guess, based on the slope.
    if f(guess) < 0.0:
        guess = guess + learning_rate
    else:
        guess = guess - learning_rate
    # Keep track of no. of iterations.
    i = i + 1
    # Break if max iterations reached.
    if i >= 1000:
        print("Max iterations reached.")
        break
# Print final values.
print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")
```

### Estimating the derivative

***

$ f(x_0) \approx \left( F(x_0 + \epsilon) - F(x_0 - \epsilon) \right) / (2 \epsilon)$

for some tiny $\epsilon$.

<br>

```python
# The derivative of F(x) at 3.0.
f(3.0)
```

```python
# Approximate the derivative of F(x) at 3.0 without using f(x).
(F(3.1) - F(2.9)) / 0.2
```

```python
# Approximate the derivative of F(x) at 3.0 without using f(x).
epsilon = 0.1
(F(3.0 + epsilon) - F(3.0 - epsilon)) / (2.0 * epsilon)
```

```python
# Create plot.
fig, ax = plt.subplots()

# Plot F(x) in 2.8 to 3.2 region - note it looks linear.
xvals = np.linspace(2.8, 3.2, 100)

# Plot F(x).
ax.plot(xvals, F(xvals), color='C2', label='F(x) between 2.8 and 3.2')

# Legend,
ax.legend();
```

```python
# Plot F(x) in 1.2 to 1.6 region - careful near features.
xvals = np.linspace(1.2, 1.6, 100)

# Create plot.
fig, ax = plt.subplots()

# Plot F(x).
ax.plot(xvals, F(xvals), color='C2', label='F(x) between 1.2 and 1.6')

# Legend,
ax.legend();
```

```python
# Function for the derivative estimate.
def fest(x, epsilon=0.001):
    return (F(x + epsilon) - F(x - epsilon)) / (2.0 * epsilon)
```

```python
# Initial guess.
guess = 3.0

# Previous guess.
prev = 2.9

# The required accuracy.
accuracy = 0.001

# The little bit to move x by.
learning_rate = 0.01

# The maximum number of iterations.
max_iter = 1000

# Set iterations count to 0.
i = 0

# Slowly decrease x, and watch the derivative of F.
while abs(guess - prev) > accuracy and (fest(guess) * fest(prev) >= 0.0):
    # Keep track of the current guess.
    prev = guess
    
    # Show the guess and the value of F(guess).
    if i % 10 == 0:
        print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")
    
    # Adjust the current guess, based on the slope.
    if ((F(guess + 0.1) - F(guess - 0.1)) / 0.2) < 0.0:
        guess = guess + learning_rate
    else:
        guess = guess - learning_rate
    
    # Keep track of no. of iterations.
    i = i + 1
    
    # Break if max iterations reached.
    if i >= max_iter:
        print("Max iterations reached.")
        break

# Print the final values.
print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")
```

### Using the derivative with the learning rate

A better idea is to multiply the learning rate by the derivative.
This changes the step size in line with the steepness of the curve.

***

```python
# Initial guess.
guess = 3.0

# Previous guess.
prev = guess

# The calculated values.
xcalc = np.array([guess])
Fcalc = np.array([F(guess)])

# The required accuracy.
accuracy = 0.0001

# The little bit to move x by.
learning_rate = 0.01

# The maximum number of iterations.
max_iter = 1000

# Set iterations count to 0.
i = 0

# Slowly decrease x, and watch the derivative of F.
while abs(fest(guess)) > accuracy:
    # Keep track of the current guess.
    prev = guess
    
    # Show the guess and the value of F(guess).
    if i % 10 == 0:
        print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")
    
    # Adjust the current guess, based on the slope.
    guess = guess - (fest(guess) * learning_rate)
    
    # Keep track of no. of iterations.
    i = i + 1
    
    # Break if max iterations reached.
    if i >= max_iter:
        print("Max iterations reached.")
        break
    
    # Append the calculated values for plotting later.
    xcalc = np.append(xcalc, [guess])
    Fcalc = np.append(Fcalc, [F(guess)])

# Final final values.
print(f"i: {i:-3}  x: {guess:4.3f}  F(x):{F(guess):6.3f}")

# Append the calculated values for plotting later.
xcalc = np.append(xcalc, [guess])
Fcalc = np.append(Fcalc, [F(guess)])

# X values
x = np.linspace(xcalc.min(), 3.0, 100)
```

```python
# x values for plot.
xvals = np.linspace(-3.0, 3.0, 100)

# Create a plot.
fig, ax = plt.subplots()

# Plot x axis.
ax.plot(xvals, xvals * 0.0, color='k', alpha=0.2, linestyle='--')

# Plot F(x).
ax.plot(xvals, F(xvals), color='C2', label='Integral')

# Plot the gradient descent values.
ax.plot(xcalc, Fcalc, 'x', color='C6', label='Descent')

# Legend.
ax.legend();
```
