```python
# Qiskit itself.

import qiskit
```

```python
# Aer provider, simulator.
import qiskit.providers.aer as aer
```

```python
# Number of qubits in total.
k = 2

# Create a circuit with k qubits and k-1 classical bits.
circuit = qiskit.QuantumCircuit(k, k-1)

# Initialize all but last qubit to 0.
for i in range(k-1):
    circuit.initialize([1, 0], i)
# Initialize last qubit to 1.
circuit.initialize([0, 1], k-1)

# Apply Hadamard gate to all qubits.
circuit.h(range(k))

# Apply U_f.
# U_f(circuit)

# Apply Hadamard to qubits 0 to k-2.
circuit.h(range(k-1))

# Measure all but last qubit.
for i in range(k-1):
    circuit.measure(i, i)

# Draw the circuit.
circuit.draw(output='mpl', scale=1.6)
```

```python
# Number of qubits in total.
k = 2

# Create a circuit with k qubits and k-1 classical bits.
circuit_a = qiskit.QuantumCircuit(k, k-1)

# Initialize all but last qubit to 0.
for i in range(k-1):
    circuit_a.initialize([1, 0], i)
# Initialize last qubit to 1.
circuit_a.initialize([0, 1], k-1)

# Apply Hadamard gate to all qubits.
circuit_a.h(range(k))

# Apply U_f.
# U_f(circuit)

# Apply Hadamard to qubits 0 to k-2.
circuit_a.h(range(k-1))

# Draw the circuit.
circuit_a.draw(output='mpl', scale=1.6)
```

```python
# Number of qubits in total.
k = 2

# Create a circuit with k qubits and k-1 classical bits.
circuit_b = qiskit.QuantumCircuit(k, k-1)

# Apply X gate to all qubits.
circuit_b.x(range(k))

# Draw the circuit.
circuit_b.draw(output='mpl', scale=1.6)
```

```python
# Join circuits.
circuit_ab = circuit_a + circuit_b
```

```python
# Draw the circuit.
circuit_ab.draw(output='mpl', scale=1.6)
```

```python
# Join circuits.
circuit_ab = circuit_a.compose(circuit_b)
```

```python
# Draw the circuit.
circuit_ab.draw(output='mpl', scale=1.6)
```
